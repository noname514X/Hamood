## 数据模拟

Python 字典，用于模拟生成带有各种问题（缺失值、异常值、重复数据、格式混乱的日期等）的数据。

```python
import pandas as pd
import numpy as np

# 模拟生成数据
data = {
    "姓名": ["张三", "李四", "王五", "赵六", "小明", "小红", "老王", "张三", "赵六"],
    "年龄": [25, 30, 22, np.nan, 29, 26, 28, 25, 200],  # 缺失值 & 异常值
    "工龄": [2, 5, 1, 8, np.nan, 3, -1, 2, 10],        # 缺失值 & 错误数据（负数）
    "薪资": [8000, 12000, 6000, 15000, 10000, 9000, 8500, 8000, 16000],  # 重复行
    "入职日期": ["2022-06-01", "2020/05/15", "2021-09-10", "2019-12-01", 
                 "2023/01/01", "2021-08-15", "2018-03-10", "2022-06-01", "2019-12-01"],  # 日期格式混乱
    "部门": ["市场部", "技术部", "市场部", "技术部", "市场部", "人事部", "技术部", "市场部", "财务部"]
}

df = pd.DataFrame(data)
print("原始数据：\n", df)

# 1. 删除重复行
df = df.drop_duplicates()
print("\n删除重复行后的数据：\n", df)

# 2. 处理缺失值
# 填充年龄和工龄的缺失值为中位数
df['年龄'].fillna(df['年龄'].median(), inplace=True)
df['工龄'].fillna(df['工龄'].median(), inplace=True)
print("\n处理缺失值后的数据：\n", df)

# 3. 处理异常值
# 替换年龄中的异常值（如200）为中位数
df.loc[df['年龄'] > 100, '年龄'] = df['年龄'].median()
# 替换工龄中的负数为0
df.loc[df['工龄'] < 0, '工龄'] = 0
print("\n处理异常值后的数据：\n", df)

# 4. 统一日期格式
df['入职日期'] = pd.to_datetime(df['入职日期'], format='%Y-%m-%d', errors='coerce')
print("\n统一日期格式后的数据：\n", df)

# 5. 检查数据类型
print("\n数据类型：\n", df.dtypes)
```

## 数据清洗
数据清洗其实就是对症下药，让数据变得干净、标准，便于后续处理。这里主要包括：

### 缺失值处理
首先看看哪些地方数据“不完整”：

```python
print("缺失值统计：\n", df.isnull().sum())
```

通过缺失值统计可以发现年龄和工龄2列分别有1个缺失值。

处理策略：

.年龄 缺失值用中位数填充（受极端值影响小）

.工龄 缺失值用部门内均值补全（让每个部门的数据更有说服力）

```python
df["年龄"]=df["年龄"].fillna(df["年龄"].median())
df["工龄"]=df["工龄"].fillna(df.groupby("部门")["工龄"].transform("mean"))
```

### 删除缺失值与填充

适用于：

缺失值占比极少（比如 < 5%）

该列不是关键数据业务

可以接受少量数据丢失

```python
import pandas as pd

# 创建数据框
df = pd.DataFrame({
    '姓名': ['张三', '李四', '王五', None, '赵六'],
    '年龄': [23, 25, None, 30, 28],
    '薪资': [5000, 7000, None, 8000, 9000]
})

# 删除含缺失值的行
df_dropna = df.dropna()

# 打印结果
print(df_dropna)
```

```python
# 使用 fillna 填充缺失值
df_filled = df.fillna({
    '姓名': '未知',
    '年龄': df['年龄'].mean(),  # 使用年龄列的平均值填充
    '薪资': 0  # 使用 0 填充薪资列的缺失值
})

# 打印结果
print(df_filled)
```

df.fillna(value) 可用于填充缺失值

df['某列'].mean()  用该列均值填充（适用于数值数据）

df['某列'].median()  用中位数填充，适用于受极端值影响的数据

df['某列'].mode()[0]  用众数填充（适用于分类变量）

```python
df_interpolated = df.interpolate()
print(df_interpolated)
```

interpolate() 采用线性插值法填充数值缺失值（适用于趋势平稳的数据）

```python
# 前向填充（用前一行数据填充）
df_ffill = df.ffill()

# 后向填充（用后一行数据填充）
df_bfill = df.bfill()

# 打印结果
print("前向填充结果：")
print(df_ffill)
print("\n后向填充结果：")
print(df_bfill)
```

### 剔除异常值

```python
import pandas as pd

# 创建数据框
df = pd.DataFrame({
    '姓名': ['张三', '李四', '王五', '赵六', '小明', '小红', '老王', '张三', '赵六'],
    '年龄': [25, 30, 22, None, 29, 26, 28, 25, 200],  # 缺失值 & 异常值
    '工龄': [2, 5, 1, 8, None, 3, -1, 2, 10],        # 缺失值 & 错误数据（负数）
    '薪资': [8000, 12000, 6000, 15000, 10000, 9000, 8500, 8000, 16000],  # 重复行
    '入职日期': ["2022-06-01", "2020/05/15", "2021-09-10", "2019-12-01", "2023/01/01", "2021-08-15", "2018-03-10", "2022-06-01", "2019-12-01"],  # 日期格式混乱
    '部门': ["市场部", "技术部", "市场部", "技术部", "市场部", "人事部", "技术部", "市场部", "财务部"]
})
df
```

```python

# 处理年龄异常值：用箱线图法（IQR）检测并剔除异常值
Q1 = df["年龄"].quantile(0.25)
Q3 = df["年龄"].quantile(0.75)
IQR = Q3 - Q1
df = df[(df["年龄"] >= Q1 - 1.5 * IQR) & (df["年龄"] <= Q3 + 1.5 * IQR)]

# 工龄负数也不合理，直接过滤
df = df[df["工龄"] >= 0]

df
```
