3.学校举办朗读比赛，邀请了10位评委为每一名参赛选手的表现打分。假设列表lst_score=[9,10,8,9,10,7,6,8,7,8],存放了某一位参赛 选⼿的所有评委打分。试编写程序，根据以下规则计算该参赛选手的最终得分： （1）去掉1个最高分 （2）去掉1个最低分 （3）最终得分为剩下8个分数的平均值

```python
lst_score = [9, 10, 8, 9, 10, 7, 6, 8, 7, 8]
sorted_scores = sorted(lst_score)[1:-1]  
final_score = sum(sorted_scores) / len(sorted_scores)
print(f"最终评分为：{final_score:.2f}")
```

4.斐波纳契数列又称黄金分割数列、兔子数列，其第1、2项均为1，从第3项开始每一项都是前两天之和，即1，1，2，3，5，8，13，21，34，…。

试编写程序，利用列表计算斐波纳契数列前30项，并输出。

```python
def fib(n):
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)

for i in range(1, 31):
    print(f"第{i}项斐波那契数列为：{fib(i)}")
```

5.甲、乙、丙、丁四人中有一人做了好事不留姓名。请大家编写程序，根据以下线索找出做好事的人。

甲：不是我

乙：是丙

丙：是丁

丁：丙说的不对

其中三人说的是真话，一人说的是假话。

```python
def check_statements(doer):

    stmt_A = (doer != 'A')      # 甲说："不是我"
    stmt_B = (doer == 'C')      # 乙说："是丙"
    stmt_C = (doer == 'D')      # 丙说："是丁"
    stmt_D = (doer != 'D')      # 丁说："丙说的不对"
    

    true_count = sum([stmt_A, stmt_B, stmt_C, stmt_D])
    

    return true_count == 3


for person in ['A', 'B', 'C', 'D']:
    if check_statements(person):
        print(f"做好事的是: {person}")
```

6.有两个列表 l1=[11,22,33] l2=[22,33,44]

a.获取内容相同的元素列表

b.获取l1中有， l2中没有的元素列表

c.获取l1和l2中内容都不同的元素

```python
l1 = [11,22,33]
l2 = [22,33,44]
lista = list(set(l1) & set(l2))
listb = list(set(l1) - set(l2))
listc = list(set(l1) ^ set(l2))
print(lista)
print(listb)
print(listc)
```

（1）需要删除的员工姓名由用户输入⼊。

（2）若用户输入的员工姓名在列表中存在，则执行删除操作；若不存在，则给出相应的提示。

（3）程序可循环执行，当用户输入姓名为“0”时，循环结束。

```python
lst_info = [['李玉','男',25],['金忠','男',23],['刘妍','女',21],['莫心','女',24],['沈冲','男',28]]
print(lst_info)
# lst_info[4][0] = '孙琉屿'
# lst_info[4][1] = '男'
# lst_info[4][2] = 20       
inputName = input('请输入需要删除的姓名：')
for i in range(len(lst_info)):
    if inputName == lst_info[i][0]:
        lst_info.remove(lst_info[i])
        break
    else:
        print('没有找到该姓名')
        break

lst_info[3][0] = '孙琉屿'
lst_info[3][1] = '男'
lst_info[3][2] = 20   
print(lst_info)
```

8.假设有字符串s=“语文:80,数学：82，英语：90，物理：85，化学：88，美术：80”，存放了某个学生的期末成绩。试编写程序， 计算该学生所有科目的总分和平均分（保留1位小数）。

```python
s = "语文:80,数学：82，英语：90，物理：85，化学：88，美术：80"

s = s.replace('：', ':').replace('，', ',')

scores = []
for pair in s.split(','):
    score = int(pair.split(':')[1])
    scores.append(score)

total = sum(scores)
average = total / len(scores)

print(f"总分：{total}")
print(f"平均分：{average:.1f}")
```
