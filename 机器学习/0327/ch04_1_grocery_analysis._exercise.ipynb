```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import os
os.chdir(r'/Users/ncc1031a/Documents/VSCode/MachineLearning/0327')
```

```python
grocery_items = set() #初始化商品集合
with open(r'grocery_dataset.txt') as f:
    reader = csv.reader(f,delimiter=',')
    for i,line in enumerate(reader):
        grocery_items.update(line)  # 集合的update方法，通过添加另一个序列中的项目来更新当前集合        
```

```python
output_list = list() # 每行为一个商品集的计数向量 
with open('grocery_dataset.txt') as f:
    reader = csv.reader(f,delimiter=',')
    for i,line in enumerate(reader):
        row_val = {item: 0 for item in grocery_items}
        row_val.update({item: 1 for item in line}) ##字典的update方法, 向字典中添加或更新键值对
        output_list.append(row_val)
```

```python
## 填空：将output_list转换为数据框 ，输出shape为 (num_transaction, total_items)
grocery_df = pd.DataFrame(output_list)
print("Shape of grocery_df:", grocery_df.shape)
```
