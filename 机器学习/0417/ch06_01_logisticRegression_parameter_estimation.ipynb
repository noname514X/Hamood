```python
import pandas as pd
import numpy as np
import os
os.chdir(r"/Users/ncc1031a/Documents/VSCode/MachineLearning/0417")
##数据读取
df = pd.read_csv("/Users/ncc1031a/Documents/VSCode/MachineLearning/0417/iris.csv")
df.head()

##选择setosa和versicolor两种
X = df.iloc[:100,[0,2]].values
y = df.iloc[:100,4].values
y = np.where(y=='setosa',0,1) ## 标签转换为0,1
X_ = np.hstack((np.ones((X.shape[0],1)),X))
w_ = np.zeros(1+X.shape[1]) 
```

```python
total_loss = []
iter = 1000
eta = 0.05
for i in range(iter):
    X_out =  X_.dot(w_)## 填空，加权求和
    y_val =  1 / (1 + np.exp(-X_out))## 填空，sigmoid函数  
    err =    y - y_val## 填空，计算误差   
    grad =   -(1.0/len(X_))*err.dot(X_)## 填空，计算梯度
    w_ =     w_ + eta * grad## 填空，更新权重
    loss = (1/len(X))*(-y.dot(np.log(y_val))-(1-y).dot(np.log(1-y_val))) ## 计算损失函数
    total_loss.append(loss)
    print('第{}轮已完成,损失为{}'.format(i+1,loss))
```

```python
#可视化
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
plt.plot(total_loss)
plt.title('Loss Function')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.show()


```
