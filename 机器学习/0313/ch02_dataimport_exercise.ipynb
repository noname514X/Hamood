1.读取我国大陆30个省、直辖市、自治区的经济发展状况数据集ecodata.xlsx，并填空

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
## 填空
data= pd.read_csv('ecodata.xlsx')
## 填空
data=data.values        #转化为数组
area=data[:,:1]
## 填空，舍弃第一列
data=data[:,1:]        #舍弃第一列
dataMat=data
## 填空，转化为浮点型
data = np.array(data, dtype=float)
np.set_printoptions(suppress=True,precision=2)
print('原始数据:\n',data)
```

```python
##填空，利用numpy计算每列标准差
temp = np.np.std(data, axis=0)
##填空，利用numpy使用数据居中处理（每列减去其均值）
data -= np.mean(data, axis=0)
data /= temp
normData=data
## 填空，计算data的转置矩阵的协方差矩阵
data = np.cov(data.T)
print('协方差矩阵:\n',data)
```

2.读入BostonHousing.csv文件，并填空

```python
import pandas as pd
import numpy as np
##填空
bostonHousing_df = pd.read_csv('boston_housing.csv')
##解释下行代码的含义 将包含空格的列名替换为更规范的格式，使其更易于使用
bostonHousing_df = bostonHousing_df.rename(columns={'CAT. MEDV': 'CAT_MEDV'})
##填空，显示数据前9行
bostonHousing_df.head(9)
```

```python
## 填空，查看数据的描述性统计结果
bostonHousing_df.describe()
```

```python
## 计算CRIM列的均值，标准差，最小值，最大值，中位数，长度，以及缺失值
## 计算CRIM列的统计量
print('Mean : ', bostonHousing_df['CRIM'].mean())
print('Std. dev : ', bostonHousing_df['CRIM'].std())
print('Min : ', bostonHousing_df['CRIM'].min())
print('Max : ', bostonHousing_df['CRIM'].max())
print('Median : ', bostonHousing_df['CRIM'].median())
print('Length : ', len(bostonHousing_df['CRIM']))
print('Number of missing values : ', bostonHousing_df['CRIM'].isnull().sum())
```

```python
## 计算所有变量的均值，标准差，最小值，最大值，中位数，长度，以及缺失值
pd.DataFrame({
    'mean': bostonHousing_df.mean(),
    'sd': bostonHousing_df.std(),
    'min': bostonHousing_df.min(),
    'max': bostonHousing_df.max(),
    'median': bostonHousing_df.median(),
    'length': bostonHousing_df.count(),
    'miss.val': bostonHousing_df.isnull().sum(),
})
```
